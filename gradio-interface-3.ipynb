```python
import gradio as gr

def greet(name):
    return f"Hello {name}"

with gr.Blocks() as demo:
    gr.Markdown("# Simple Gradio App")
    
    name = gr.Textbox(label="Enter your Name")
    output = gr.Textbox(label="Output")
    
    btn = gr.Button("Submit")
    btn.click(greet, inputs=name, outputs=output)

demo.launch()
```
