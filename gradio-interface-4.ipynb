```python
import gradio as gr
from PIL import Image

def convert_to_grayscale(img):
    if img is None:
        return None
    gray = img.convert("L")   # Convert to grayscale
    return gray

# Create Gradio Interface
demo = gr.Interface(
    fn=convert_to_grayscale,
    inputs=gr.Image(type="pil"),
    outputs=gr.Image(type="pil"),
    title="Grayscale Image Converter",
    description="Upload an image and convert it to grayscale"
)

demo.launch()
```
